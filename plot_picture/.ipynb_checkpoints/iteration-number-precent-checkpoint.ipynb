```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from matplotlib.ticker import FuncFormatter
```

```python
x_iter_number=[30,40,50,60,70,80]
y_Amazon=[0.005,0.123,0.444,0.57,0.664,0.71]
y_Henan=[0.393,0.553,0.695,0.787,0.839,0.875]
y_Elunchun=[0.463,0.687,0.828,0.911,0.951,0.969]
```

```python
def to_percent(temp, position):
    return '%1.0f'%(100*temp) + '%'

fig = plt.figure(figsize=(6,4),dpi=300)
plt.rcParams['axes.labelsize'] = 10.5
plt.plot(x_iter_number,y_Amazon,linewidth=3, color='mediumspringgreen', \
#          marker='o', linestyle='solid', \
         label="Amazon")
plt.plot(x_iter_number,y_Henan,linewidth=3,color='cornflowerblue', \
#          marker='.', linestyle='-.', \
         label="Henan")
plt.plot(x_iter_number,y_Elunchun,linewidth=3,color='coral',\
#          marker='1', linestyle=':', \
         label="Elunchun")
plt.legend(loc='upper left',prop={'size':9})
plt.xlabel('Number of iterations')
plt.ylabel('Precent')
# fmt = '%.0f%' # Format you want the ticks, e.g. '40%'

# yticks = mtick.FormatStrFormatter(fmt)

# plt.yaxis.set_major_formatter(yticks)
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_percent))
# plt.gca().grid(axis='y',linestyle='-.') # 打开网格线
# plt.show()
# plt.savefig('D:/Desktop/mypaper/picture/iteration-precent.jpg',bbox_inches='tight',pad_inches=0.02)
```
