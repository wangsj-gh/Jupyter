```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

from scipy.stats import norm
from osgeo import gdal
import os
import time
import sys

from scipy.optimize import fsolve
```

```python
class Dataset:
    def __init__(self, in_file):
        self.in_file = in_file  # Tiff或者ENVI文件

        dataset = gdal.Open(self.in_file)
        self.XSize = dataset.RasterXSize  # 网格的X轴像素数量
        self.YSize = dataset.RasterYSize  # 网格的Y轴像素数量
        self.Bands = dataset.RasterCount  # 波段数
        self.GeoTransform = dataset.GetGeoTransform()  # 投影转换信息
        self.ProjectionInfo = dataset.GetProjection()  # 投影信息
    
    def get_data(self):
        #band: 读取第几个通道的数据
        dataset = gdal.Open(self.in_file)
        data = dataset.ReadAsArray(0,0,self.XSize,self.YSize)
        return data
    

    def get_lon_lat(self):
        #获取经纬度信息
        gtf = self.GeoTransform
        x_range = range(0, self.XSize)
        y_range = range(0, self.YSize)
        x, y = np.meshgrid(x_range, y_range)
        lon = gtf[0] + x * gtf[1] + y * gtf[2]
        lat = gtf[3] + x * gtf[4] + y * gtf[5]
        
        lon_lat=[]
        for (longitude,latitude) in zip(lon,lat):
            lon_lat.append(list(zip(longitude,latitude)))
            
        return np.array(lon_lat)
    
    def new_dataset(self,data,lon_lat):
        new_dataset=[]
        for i in range(self.YSize):
            for j in range(self.XSize):
                x1 = lon_lat[i,j,:]
                x2 = data[i,j,:]
                x=np.hstack((x1,x2))
                new_dataset.append(x)
            
        return np.array(new_dataset)
    
```

```python
def normfun(x, mu, sigma):
    pdf = np.exp(-((x - mu) ** 2) / (2 * sigma ** 2)) / (sigma * np.sqrt(2 * np.pi))
    return pdf

def Doublelogic(x, m1, m2, m3, m4, m5, m6):
    return m1 + m2 /(1 + np.exp(-m3 * (x-m4))) - m2/(1 + np.exp(-m5 * (x-m6)))

#注意初值
def fitting(yData, xData) :
    Parameters, pcov = curve_fit(Doublelogic, xData, yData, p0=[5,40,0.1,140,0.1,270], maxfev=10000000)#,method='trf', maxfev=1000000)
    return Parameters

def for_map(yData, xData):
    result=np.array(list(map(fitting,yData,xData)))
 
    return result
```

矩估计方法


```python
normal_mu  = error.mean() # 获得数据集的平均值
normal_sigma  = ((error-normal_mu)**2).mean()   # 获得数据集的标准差
normal_B3=((error-normal_mu)**3).mean()
# print(normal_mu,normal_sigma,normal_B3)


# get solution by fsolve
def get_p(p):
    
    func = normal_B3**2/(4*(np.pi-3)*((1-p)*np.sqrt(p)-p*np.sqrt(1-p))+\
         (np.pi-4)*(p*np.sqrt(p)-(1-p)*np.sqrt(1-p)))**2-\
        2*normal_sigma**3/(np.pi-2+(4-np.pi)*(np.sqrt(p*(1-p))))**3
    return func

def solve_p(value):
    p=value[0]
    return [get_p(p)]

# get solution by newton

def diff(p):
    par1=p+0.00001
    par2=p-0.00001
    y1=get_p(par1)
    y2=get_p(par2)
    deriv=(y1-y2)/0.00002
    return deriv

def newton(convergence, cur_root):
    
    for i in range(100):
        fun = get_p(cur_root)
        dif_func = diff(cur_root)
        cur_root = cur_root - fun / dif_func
        if abs(fun)<convergence:
            break
    return cur_root
```

```python
solved_fsolve_p=fsolve(solve_p,[0.6])
solved_newton_p=newton(0.001,0.6)
print(solved_fsolve_p,solved_newton_p)
```

极大似然估计方法


```python
def sign1(x,u):
    total=[]
    for i in range(len(x)):
        if x[i]<=u:
            a=1
        else:
            a=0
        total.append(a)
    return total

def sign2(x,u):
    total=[]
    for i in range(len(x)):
        if x[i]>u:
            a=1
        else:
            a=0
        total.append(a)
    return total

    
def sum_mu(p,u,sigma):

    aa=(1-p)*(x-u)*sign1(x,u)+p*(x-u)*sign2(x,u) 
    total=sum(aa)
    return 2*total/sigma**2

def sum_sigma(p,u,sigma):

    aa=(1-p)*(x-u)**2*sign1(x,u)+p*(x-u)**2*sign2(x,u)
    total=sum(aa)
    return -n/sigma+2*total*sigma**(-3)

def sum_p(p,u,sigma):

    aa=(x-u)**2*sign1(x,u)-(x-u)**2*sign2(x,u)
    total=sum(aa)
    npn=n*(np.sqrt(1-p)-p*np.sqrt(1-p)-p*np.sqrt(p))/(2*p*(1-p)*(np.sqrt(p)+np.sqrt(1-p)))
    return npn+total/sigma**2

def solve_function(value):
    p,u,sigma=value[0],value[1],value[2]
    return [
        sum_mu(p,u,sigma),
        sum_sigma(p,u,sigma),
        sum_p(p,u,sigma),
    ]
```
