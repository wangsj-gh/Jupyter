```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from matplotlib.ticker import FuncFormatter
```

```python
# python
laitime_python=[0.187,4.31,136.26,5718]
ndvitime_python=[0.259,2.67,107.08,1318]
evitime_python=[0.32,2.21,107.31,1320]

#excel
laitime_excel=[1.07,110.5,3538.37,83195]
ndvitime_excel=[2.64,51.2,2927.16,68825]
evitime_excel=[0.95,150.4,2850.19,67009]

#gee
laitime_gee=[42.471,60,239,2400]
ndvitime_gee=[35.975,54,120,1080]
evitime_gee=[35.272,56,120,1080]

def formatnum(x, pos):
    return '$%.1f$x$10^{4}$' % (x/1000)


def plot_line(laitime,ndvitime,evitime,y_name,ax1_1,chose):
    # mediumspringgreen，cornflowerblue，coral，aqua，
    labels = ['1', '156','12656','297571']
    x = [0.45,1.45,2.45,3.45] 
    ax1_1.plot(x,laitime,linewidth=3, color='mediumspringgreen', marker='o', linestyle='solid', label="LAI")
    ax1_1.plot(x,ndvitime,linewidth=3,color='cornflowerblue', marker='.', linestyle='-.', label="NDVI")
    ax1_1.plot(x,evitime,linewidth=3,color='coral', marker='1', linestyle=':', label="EVI")
#     ax1_1.set_xlim(0,3)
    if chose:
        ax1_1.legend()
        
    plt.xticks(x, labels)
    
#     ax1_1.yaxis.set_major_formatter(FuncFormatter(formatnum))
    ax1_1.yaxis.get_major_formatter().set_powerlimits((0,1))
    ax1_1.set_ylabel('Time(s)')
    ax1_1.set_xlabel('Number of points')
    ax1_1.grid(axis='y',linestyle='-.') # 打开网格线
    ax1_2 = ax1_1.twinx()
    ax1_2.set_ylabel(y_name)
    ax1_2.yaxis.set_major_locator(plt.NullLocator()) # 删除坐标轴的刻度显示

    
fig = plt.figure(figsize=(7,9.5),dpi=300)
plt.rcParams['axes.labelsize'] = 10.5 # xy轴label的size
# plt.rcParams['xtick.labelsize'] = 8  # x轴ticks的size
# plt.rcParams['ytick.labelsize'] = 15  # y轴ticks的size

ax1_1 = fig.add_subplot(3,1,1)
plot_line(laitime_excel,ndvitime_excel,evitime_excel,'Excel',ax1_1,1)

ax1_2 = fig.add_subplot(3,1,2)
plot_line(laitime_python,ndvitime_python,evitime_python,'Python',ax1_2,1)

ax1_3 = fig.add_subplot(3,1,3)
plot_line(laitime_gee,ndvitime_gee,evitime_gee,'GEE',ax1_3,1)

plt.subplots_adjust(wspace=0,hspace=0.21)

# plt.show()
# plt.savefig('D:/Desktop/mypaper/picture/all-time-1-156-12656-297571.jpg',bbox_inches='tight',pad_inches=0.02)
```

```python
#items
geeitem_12656 = [34,45,40]
pythonitem_12656 = [149,108,122]
excelitem_12656 = [28,22,22]
    
def plot_bar(geeitem,pythonitem,excelitem,text_name,ax1_2):
    width = 0.3  # 柱形的宽度
    x1_list = []
    x2_list = []
    x3_list = []
    labels = ['LAI', 'NDVI','EVI']
    x = [0.45,1.45,2.45] 
    for i in range(3):
        x1_list.append(i)
        x2_list.append(i+width)
        x3_list.append(i+2*width)
    ax1_2.bar(x1_list,geeitem, color='mediumspringgreen',width=width, align='edge', label="GEE")
    ax1_2.bar(x2_list,pythonitem,color='cornflowerblue', width=width, align='edge', label="Python")
    ax1_2.bar(x3_list,excelitem,color='coral',width=width, align='edge', label="Excel")
    plt.xticks(x, labels) # 刻度值
    # ax1_2.grid(axis='y',linestyle='-.') # 打开网格线
    ax1_2.set_ylabel('Items')
    ax1_2.legend()
#     ax1_2.text(0,0,'(%s)'%text_name,fontsize=14)

fig = plt.figure(figsize=(15,15),dpi=300)
plt.rcParams['axes.labelsize'] = 10.5 # xy轴label的size
ax1_1 = fig.add_subplot(3,2,1)
plt.ylim(0,170)
plot_bar(geeitem_12656,pythonitem_12656,excelitem_12656,"c",ax1_1)
# plt.savefig('D:/Desktop/mypaper/picture/all-item-12656.jpg',bbox_inches='tight',pad_inches=0.02)
```
