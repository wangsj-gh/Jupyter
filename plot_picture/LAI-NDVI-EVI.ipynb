```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'D:/Desktop/mypaper/point/paperdata.xlsx'
LAIresult=pd.read_excel(path,index_col='day',parse_dates=True,sheet_name = 'LAI-result')
NDVIresult=pd.read_excel(path,index_col='day',parse_dates=True,sheet_name = 'NDVI-result')
EVIresult=pd.read_excel(path,index_col='day',parse_dates=True,sheet_name = 'EVI-result')
```

```python
LAIx=LAIresult.index
LAIy1=LAIresult['data']
# LAIy2=LAIresult['solver']
LAIy3=LAIresult['LM']
# LAIy4=LAIresult['python']
```

```python
NDVIx=NDVIresult.index
NDVIy1=NDVIresult['data']
# NDVIy2=NDVIresult['solver']
NDVIy3=NDVIresult['LM']
# NDVIy4=NDVIresult['python']
```

```python
EVIx=EVIresult.index
EVIy1=EVIresult['data']
# EVIy2=EVIresult['solver']
EVIy3=EVIresult['LM']
# EVIy4=EVIresult['python']
```

```python
# 设置figure_size尺寸
plt.rcParams['figure.figsize'] = (8.0, 10.0) 

#图片像素
plt.rcParams['savefig.dpi'] = 300 

#分辨率
plt.rcParams['figure.dpi'] = 500 
```

```python
legendsize=10
pointsize=20
plt.figure(figsize=(8,10),dpi=300)
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['xtick.labelsize'] = 6  # x轴ticks的size
plt.rcParams['ytick.labelsize'] = 15  # y轴ticks的size
plt.subplot(3,1,1)
plt.plot(LAIx,LAIy3,color='mediumspringgreen',linewidth=3.0,label='GEE')
# plt.scatter(LAIx,LAIy2,marker='^',color='green',s=pointsize,label='solver')
plt.scatter(LAIx,LAIy1,marker='^',color='black',s=pointsize,label='LAI')
# plt.scatter(LAIx,LAIy4,marker='*',color='m',s=pointsize,label='python')
p0=0.197320
p1=3.384014
p2=0.068320
p3=126.530350
p4=0.071226
p5=283.7165
plt.annotate(r'$y(t)=%s+\frac{%s}{1+e^{-%s(t-%s)}}+\frac{%s}{1+e^{-%s(t-%s)}}$'%(p0,p1,p2,p3,p1,p4,p5),
#              xy=(152,3),xycoords='data',xytext=(15,3.5),
             xy=(152,6),xycoords='data',xytext=(20,5.4),
             fontsize=10)
#              arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-0.05',color='red'))

plt.xlim((1,370))
plt.ylim((0,6))
plt.xlabel('Day')
plt.ylabel('LAI')
plt.text(6,5.4,'(a)')
plt.legend(loc='upper right',prop={'size':legendsize})

plt.subplot(3,1,2)
plt.plot(NDVIx,NDVIy3,color='cornflowerblue',linewidth=3.0,label='GEE')
# plt.scatter(NDVIx,NDVIy2,marker='^',color='green',s=pointsize,label='solver')
plt.scatter(NDVIx,NDVIy1,marker='8',color='black',s=pointsize,label='NDVI')
# plt.scatter(NDVIx,NDVIy4,marker='*',color='m',s=pointsize,label='python')

p0=0.045488
p1=0.766361
p2=0.055246
p3=92.321457
p4=0.025295
p5=337.3139
plt.annotate(r'$y(t)=%s+\frac{%s}{1+e^{-%s(t-%s)}}+\frac{%s}{1+e^{-%s(t-%s)}}$'%(p0,p1,p2,p3,p1,p4,p5),
#              xy=(152,0.77),xycoords='data',xytext=(15,0.85),
             xy=(152,0.77),xycoords='data',xytext=(20,0.9),
             fontsize=10)
#              arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-0.05',color='red'))

plt.xlim((1,370))
plt.ylim((0,1))
plt.xlabel('Day')
plt.ylabel('NDVI')
plt.text(6,0.9,'(b)')
plt.legend(loc='upper right',prop={'size':legendsize})

plt.subplot(3,1,3)
plt.plot(EVIx,EVIy3,color='coral',linewidth=3.0,label='GEE')
# plt.scatter(EVIx,EVIy2,marker='^',color='green',s=pointsize,label='solver')
plt.scatter(EVIx,EVIy1,marker='s',color='black',s=pointsize,label='EVI')
# plt.scatter(EVIx,EVIy4,marker='*',color='m',s=pointsize,label='python')

p0=0.078029
p1=0.548960
p2=0.059740
p3=110.739128
p4=0.023726
p5=280.5085
plt.annotate(r'$y(t)=%s+\frac{%s}{1+e^{-%s(t-%s)}}+\frac{%s}{1+e^{-%s(t-%s)}}$'%(p0,p1,p2,p3,p1,p4,p5),
#              xy=(152,0.55),xycoords='data',xytext=(15,0.6),
                xy=(152,0.55),xycoords='data',xytext=(20,0.71),
             fontsize=10)
#              arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=-0.05',color='red'))

plt.xlim((1,370))
plt.ylim((0,0.8))
plt.xlabel('Day')
plt.ylabel('EVI')
plt.text(6,0.71,'(c)')
plt.legend(loc='upper right',prop={'size':legendsize})

# plt.savefig('D:/Desktop/mypaper/picture/LAI-NDVI-EVI.jpg',bbox_inches='tight',pad_inches=0.1)
```
