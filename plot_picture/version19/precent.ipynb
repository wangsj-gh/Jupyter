```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker
```

```python
x=[10,20,30,40,50,60,70,80]
nls=[7.62777E-05,0.026849733,0.133180778,0.296720061,0.437604882,0.544774981,0.597177727,0.625171625]
lm=[0,0.234074491,0.681818182,0.798198705,0.849610658,0.87911624,0.888451074,0.891922319]
```

```python
# 设置figure_size尺寸
plt.rcParams['figure.figsize'] = (10,6) 

#图片像素
plt.rcParams['savefig.dpi'] = 500 

#分辨率
plt.rcParams['figure.dpi'] = 500 

# plt.rcParams['axes.labelsize'] = 10.5

# x轴ticks的size
plt.rcParams['xtick.labelsize'] = 15

# y轴ticks的size
plt.rcParams['ytick.labelsize'] = 15  

#设置图例并且设置图例的字体及大小
legend_size = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 15,
}

xylable_size = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 20,
}
```

```python
fig, ax = plt.subplots()
ax.plot(x,lm,linewidth=3, linestyle='solid', color='mediumspringgreen',label="LM")
ax.plot(x,nls,linewidth=3, linestyle='solid',color='cornflowerblue', label="NLS")

ax.legend(loc='upper left',prop=legend_size)
ax.set_ylabel('Rate of convergence',xylable_size)
ax.set_xlabel('Number of iterations',xylable_size)
ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals=0))
# plt.show()
plt.savefig('D:/Desktop/mypaper/version19/picture/precent.jpg',bbox_inches='tight',pad_inches=0.02)
```
