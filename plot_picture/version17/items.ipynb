```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from matplotlib.ticker import FuncFormatter
# from collections import OrderedDict
from matplotlib.pyplot import MultipleLocator
```

```python
#items
geeitem_12656 = [34,45,40]
geestd_12656=[18.13,24.57,22.39]

pythonitem_12656 = [149,122,109]
pythonstd_12656=[219,198.28,54.64] 

excelitem_12656 = [28,22,22]
excelstd_12656=[10.62,7.58,4.96]
    
def plot_bar(geeitem,geestd,pythonitem,pythonstd,excelitem,excelstd,text_name,ax1_2):
    width = 0.25  # 柱形的宽度
    x1_list = []
    x2_list = []
    x3_list = []
    labels = ['LAI', 'NDVI','EVI']
    x = [0.39,1.39,2.39] 
    for i in range(3):
        x1_list.append(i)
        x2_list.append(i+width+0.02)
        x3_list.append(i+2*width+0.04)
    ax1_2.bar(x1_list,geeitem, color='mediumspringgreen',edgecolor='black',linewidth=0.5,\
              yerr = geestd, error_kw = {'ecolor' : '0.4','capsize' :2},\
              width=width, align='edge', label="LM")
    ax1_2.bar(x2_list,pythonitem,color='cornflowerblue',edgecolor='black',linewidth=0.5, \
              yerr = pythonstd, error_kw = {'ecolor' : '0.4', 'capsize' :2},\
              width=width, align='edge', label="LS")
    ax1_2.bar(x3_list,excelitem,color='coral',edgecolor='black',linewidth=0.5,\
              yerr = excelstd, error_kw = {'ecolor' : '0.4', 'capsize' :2},\
              width=width, align='edge', label="GRG")
    
#     for i in range(len(x1_list)):
#         plt.text(x1_list[i]+0.15,geeitem[i],'%s' %geeitem[i],ha='center')
#         plt.text(x2_list[i]+0.15,pythonitem[i],'%s' %pythonitem[i],ha='center')
#         plt.text(x3_list[i]+0.15,excelitem[i],'%s' %excelitem[i],ha='center')

    y_major_locator=MultipleLocator(50)
    ax1_2.yaxis.set_major_locator(y_major_locator)
    plt.xticks(x, labels) # 刻度值
    
#     ax1_2.grid(axis='y',linestyle='--') # 打开网格线

    ax1_2.set_ylabel('Number of iterations')
    ax1_2.legend(loc='upper right')
    
    
#     ax1_2.text(0,0,'(%s)'%text_name,fontsize=14)

fig = plt.figure(figsize=(15,15),dpi=500)
plt.rcParams['axes.labelsize'] = 10.5 # xy轴label的size
ax1_1 = fig.add_subplot(3,2,1)
plt.ylim(-100,400)
plot_bar(geeitem_12656,geestd_12656,pythonitem_12656,pythonstd_12656,\
         excelitem_12656,excelstd_12656,"c",ax1_1)
plt.savefig('D:/Desktop/mypaper/version17/picture/all-item-12656-1.jpg',bbox_inches='tight',pad_inches=0.02)
```
