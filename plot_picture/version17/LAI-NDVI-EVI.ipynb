```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = r'D:\Desktop\mypaper\version17\data\point\total_data.xlsx'
LAIresult=pd.read_excel(path,sheet_name = 'lai',header=0,names=['name', 'value'])
NDVIresult=pd.read_excel(path,sheet_name = 'ndvi',header=0,names=['name', 'value'])
EVIresult=pd.read_excel(path,sheet_name = 'evi',header=0,names=['name', 'value'])
```

```python
LAIx=LAIresult.name
LAIy1=LAIresult.value/10
```

```python
NDVIx=NDVIresult.name
NDVIy1=NDVIresult.value/10000
```

```python
EVIx=EVIresult.name
EVIy1=EVIresult.value/10000
```

```python
# 设置figure_size尺寸
plt.rcParams['figure.figsize'] = (8.0, 10.0) 

#图片像素
plt.rcParams['savefig.dpi'] = 300 

#分辨率
plt.rcParams['figure.dpi'] = 500 
```

```python
legendsize=8
pointsize=20

plt.figure(figsize=(8,10),dpi=300)
plt.rcParams['axes.labelsize'] = 10.5
# plt.rcParams['xtick.labelsize'] = 6  # x轴ticks的size
# plt.rcParams['ytick.labelsize'] = 15  # y轴ticks的size

plt.subplot(3,1,1)
p0=0.181
p1=4.176
p2=0.074
p3=155.3
p4=0.039
p5=252.4

LAIy2=p0+p1/(1+np.exp(p2*p3-p2*LAIx))-p1/(1+np.exp(p4*p5-p4*LAIx))
plt.plot(LAIx,LAIy2,color='mediumspringgreen',linewidth=3.0,\
         label=r'$%s+\frac{%s}{1+e^{-%s(t-%s)}}-\frac{%s}{1+e^{-%s(t-%s)}}$'%(p0,p1,p2,p3,p1,p4,p5))
plt.scatter(LAIx,LAIy1,marker='^',color='black',s=pointsize,label='LAI')
plt.xlim((1,370))
plt.ylim((0,5))
plt.xlabel('Day')
plt.ylabel('LAI')
plt.text(6,4.6,'(a)')
plt.legend(loc='upper right',prop={'size':legendsize})

plt.subplot(3,1,2)
p0=0.095
p1=0.735
p2=0.075
p3=124.7
p4=0.035
p5=308.6
NDVIy2=p0+p1/(1+np.exp(p2*p3-p2*NDVIx))-p1/(1+np.exp(p4*p5-p4*NDVIx))

plt.plot(NDVIx,NDVIy2,color='cornflowerblue',linewidth=3.0,\
         label=r'$%s+\frac{%s}{1+e^{-%s(t-%s)}}-\frac{%s}{1+e^{-%s(t-%s)}}$'%(p0,p1,p2,p3,p1,p4,p5))
plt.scatter(NDVIx,NDVIy1,marker='8',color='black',s=pointsize,label='NDVI')
plt.xlim((1,370))
plt.ylim((0,1.2))
plt.xlabel('Day')
plt.ylabel('NDVI')
plt.text(6,1.1,'(b)')
plt.legend(loc='upper right',prop={'size':legendsize})

plt.subplot(3,1,3)
p0=0.109
p1=1.082
p2=0.042
p3=152.1
p4=0.022
p5=201.9
EVIy2=p0+p1/(1+np.exp(p2*p3-p2*EVIx))-p1/(1+np.exp(p4*p5-p4*EVIx))

plt.plot(EVIx,EVIy2,color='coral',linewidth=3.0,\
         label=r'$%s+\frac{%s}{1+e^{-%s(t-%s)}}-\frac{%s}{1+e^{-%s(t-%s)}}$'%(p0,p1,p2,p3,p1,p4,p5))
plt.scatter(EVIx,EVIy1,marker='s',color='black',s=pointsize,label='EVI')
plt.xlim((1,370))
plt.ylim((0,0.7))
plt.xlabel('Day')
plt.ylabel('EVI')
plt.text(6,0.64,'(c)')
plt.legend(loc='upper right',prop={'size':legendsize})

plt.savefig('D:/Desktop/mypaper/version17/picture/LAI-NDVI-EVI1.jpg',bbox_inches='tight',pad_inches=0.02)
```
