```python
# point
geetime = [42.471,49.975,35.272]
pythontime = [0.187,0.259,0.32]
exceltime = [1.07,2.64,0.95]

geeitem = [44,56,39]
pythonitem = [79,103,58]
excelitem = [17,18,20]
# ax1_1.set_ylim(0,55)

#156 point
geetime_156 = [100,120,120]
pythontime_156 = [67.7,12.9,22]
exceltime_156 = [110.5,51.2,150.4]

geeitem_156 = [60,65,65]
pythonitem_156 = [131,94,98]
excelitem_156 = [19,24,24]
# ax1_1.set_ylim(0,160)

#12656 point
geetime_12656 = [420,120,300]
pythontime_12656 = [5556.86,3923.23,14648.24]
exceltime_12656 = [3538.37,5904.64,2850.19]

geeitem_12656 = [60,60,65]
pythonitem_12656 = [175,131,122]
excelitem_12656 = [28,22,22]



def plot_line(geetime,pythontime,exceltime,y_max_lim,text_name,ax1_1):
    # mediumspringgreen，cornflowerblue，coral，aqua，
    # fig = plt.figure(figsize=(10,4),dpi=300)
    labels = ['LAI', 'NDVI','EVI']
    x = [0.45,1.45,2.45] 
#     ax1_1 = fig.add_subplot(1,2,1)
    ax1_1.plot(x,geetime, color='mediumspringgreen', marker='o', linestyle='solid', label="GEE")
    ax1_1.plot(x,pythontime,color='cornflowerblue', marker='o', linestyle='solid', label="Python")
    ax1_1.plot(x,exceltime,color='coral', marker='o', linestyle='solid', label="Excel")
    ax1_1.legend()
    plt.xticks(x, labels) 
    ax1_1.set_xlim(0,3)
    ax1_1.set_ylim(0,y_max_lim)
    ax1_1.set_ylabel('Time(s)')
    ax1_1.grid(axis='y',linestyle='-.') # 打开网格线

    # ax1_1.xaxis.set_major_locator(plt.NullLocator()) # 删除坐标轴的刻度显示
    # plt.subplots_adjust(wspace=0,hspace=0.05)
    ax1_1.text(0,0,'(%s)'%text_name,fontsize=14)

    
def plot_bar(geeitem,pythonitem,excelitem,text_name,ax1_2):
    width = 0.3  # 柱形的宽度
    x1_list = []
    x2_list = []
    x3_list = []
    x = [0.45,1.45,2.45] 
    labels = ['LAI', 'NDVI','EVI']
    for i in range(3):
        x1_list.append(i)
        x2_list.append(i+width)
        x3_list.append(i+2*width)
    ax1_2.bar(x1_list,geeitem, color='mediumspringgreen',width=width, align='edge', label="GEE")
    ax1_2.bar(x2_list,pythonitem,color='cornflowerblue', width=width, align='edge', label="Python")
    ax1_2.bar(x3_list,excelitem,color='coral',width=width, align='edge', label="Excel")
    plt.xticks(x, labels) # 刻度值
    # ax1_2.grid(axis='y',linestyle='-.') # 打开网格线
    ax1_2.set_ylabel('Item')
    ax1_2.legend()
    ax1_2.text(0,0,'(%s)'%text_name,fontsize=14)
    
    
fig = plt.figure(figsize=(12,15),dpi=300)

ax1_1 = fig.add_subplot(3,2,1)
plot_line(geetime,pythontime,exceltime,55,'A',ax1_1)

ax1_2 = fig.add_subplot(3,2,2)
plot_bar(geeitem,pythonitem,excelitem,"a",ax1_2)

ax1_3 = fig.add_subplot(3,2,3)
plot_line(geetime_156,pythontime_156,exceltime_156,160,'B',ax1_3)

ax1_4 = fig.add_subplot(3,2,4)
plot_bar(geeitem_156,pythonitem_156,excelitem_156,"b",ax1_4)



ax1_5 = fig.add_subplot(3,2,5)
plot_line(geetime_12656,pythontime_12656,exceltime_12656,16000,'C',ax1_5)

ax1_6 = fig.add_subplot(3,2,6)
plot_bar(geeitem_12656,pythonitem_12656,excelitem_12656,"c",ax1_6)


# 设置柱形的间隔

# ax1.plot(x,geetime, color='red', marker='o', linestyle='solid', label="GEE")
# ax1.plot(x,pythontime,color='blue', marker='o', linestyle='solid', label="Python")
# ax1.plot(x,exceltime,color='g', marker='o', linestyle='solid', label="Excel")

# ax1.legend()
# ax1.set_ylim(32, 55) 
# ax1_1.set_ylim(0, 3)

# ax1.grid(axis='both',linestyle='-.') # 打开网格线
# ax1_1.grid(axis='y',linestyle='-.')   # 打开网格线


# ax1_1.spines['top'].set_visible(False)    # 边框控制
# ax1_1.spines['bottom'].set_visible(True) # 边框控制
# ax1_1.spines['right'].set_visible(False)  # 边框控制

# x1_2.spines['top'].set_visible(False)   # 边框控制
# x1_2.spines['bottom'].set_visible(False) # 边框控制
# x1_2.spines['right'].set_visible(False)  # 边框控制

# 绘制断层线
# d = 0.01  # 断层线的大小
# kwargs = dict(transform=ax1.transAxes, color='k', clip_on=False)
# ax1.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal

# kwargs.update(transform=ax1_1.transAxes, color='k')  # switch to the bottom axes
# ax1_1.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal


# plt.show()
# plt.savefig('D:/Desktop/mypaper/picture/all-time-item-1-156-12656.jpg',bbox_inches='tight',pad_inches=0.1)
```
