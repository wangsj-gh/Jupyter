```python
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.arange(0, 2, 0.1)
y= np.exp(-x)/x
plt.plot(x, y, label="sigmoid")

plt.xlabel("x")
plt.ylabel("y")
# plt.ylim(0, 1)
plt.legend()
plt.show()
```

```python
from scipy import integrate
def f(x):
    return np.exp(-x)/x
x = np.arange(1, 2, 0.1)
y=[]
for i in x:
    v, err = integrate.quad(f, 1, i)
    y.append(v)
plt.plot(x, y, label="sigmoid")
plt.show()
```
