```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
def function(x):
    p0,p1,p2,p3,p4,p5=10,40,0.05,120,0.06,270
    return p0+(p1)/(1+np.exp(p2*p3-p2*x))-(p1)/(1+np.exp(p4*p5-p4*x))
def spring(x):
    p0,p1,p2,p3,p4,p5=10,40,0.05,120,0.06,270
    return p0+(p1)/(1+np.exp(p2*p3-p2*x))
def autumn(x):
    p0,p1,p2,p3,p4,p5=10,40,0.05,120,0.06,270
    return -(p1)/(1+np.exp(p4*p5-p4*x))
```

```python
plt.figure(figsize=(10,2),dpi=300)
plt.subplot(1,3,1)
x=np.arange(1, 365)
y=function(x)
plt.plot(x,y,label='Double')
plt.legend(loc='upper right',prop={'size':7})

plt.subplot(1,3,2)
x=np.arange(1, 365)
y=spring(x)
plt.plot(x,y,label='Spring')
plt.legend(loc='upper left',prop={'size':7})

plt.subplot(1,3,3)
x=np.arange(1, 365)
y=autumn(x)
plt.plot(x,y,label='Autumn')
plt.legend(loc='upper right',prop={'size':7})

plt.show()
```
